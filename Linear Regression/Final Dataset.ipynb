```python
import pandas as pd
```

```python
url1 ="https://raw.githubusercontent.com/polinaprinii/DAP-Project/main/Cases_and_Death_Rates/Data/Data%20for%20Multiple%20Linear%20Regression/Global_and_US_Confirmed_Cases_And_Deaths.csv?token=ASV4RS6WZYFUWQ325ZCQI3TBXDRT6"
cases_deaths =pd.read_csv(url1)

url2 = "https://raw.githubusercontent.com/polinaprinii/DAP-Project/main/Restriction%20and%20Testing%20Analysis/Restrictions/restrictions.csv?token=ASV4RS2EE7HC7DMD6OFAR4DBW6TQ6"
restrictions = pd.read_csv(url2)

url3 = "https://raw.githubusercontent.com/polinaprinii/DAP-Project/main/Restriction%20and%20Testing%20Analysis/world%20testing/testing.csv?token=ASV4RS424SAEGIM3KBA7P23BW6TTM"
tests = pd.read_csv(url3)

url4 = "https://raw.githubusercontent.com/polinaprinii/DAP-Project/main/Vaccine_Analysis_by_Country/Main%20Project/CSV%20Files/finalexport.csv?token=ASV4RSZSY4LEVQIIVPRB43LBW6WDI"
vaccine = pd.read_csv(url4)
```

```python
merge1 = pd.merge(restrictions, tests, on ="isocode")
merge1 = merge1[['isocode','country_x','MaxStringencyIndex','MeanStringencyIndex','MaxFacialCoverings','MeanFacialCoverings','MaxStayHome','MeanStayHome','tests']]
merge1
```

```python
merge2 = pd.merge(merge1,cases_deaths, on="isocode")
merge2 = merge2[['isocode','MaxStringencyIndex','MeanStringencyIndex','MaxFacialCoverings','MeanFacialCoverings','MaxStayHome','MeanStayHome','tests','country','Confirmed','Deaths']]
merge2
```

```python
merge3 = pd.merge(merge2, vaccine, on ="isocode")
merge3
```

```python
merge3 = merge3[['isocode','MaxStringencyIndex','MeanStringencyIndex','MaxFacialCoverings','MeanFacialCoverings','MaxStayHome','MeanStayHome','tests','country_x','Confirmed','Deaths']]
merge3
```

```python
finalmerge = merge3[['isocode','country_x','Confirmed','Deaths','tests','MaxStringencyIndex','MeanStringencyIndex','MaxFacialCoverings','MeanFacialCoverings','MaxStayHome','MeanStayHome']]
finalmerge = pd.merge(finalmerge, vaccine, on='isocode')
finalmerge = finalmerge[['isocode','country_x','Confirmed','Deaths','tests','MaxStringencyIndex','MeanStringencyIndex','MaxFacialCoverings','MeanFacialCoverings','MaxStayHome','MeanStayHome','totalvaccinations','totalbooster','fullyvaccinated']]
finalmerge
```

```python
finalmerge = finalmerge.rename(columns = {'country_x':'country'})
```
